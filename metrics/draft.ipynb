```python
from subgrounds.subgrounds import Subgrounds
from subgrounds.subgraph import SyntheticField
from datetime import datetime
from subgrounds.plotly_wrappers import Bar, Figure
from subgrounds.dash_wrappers import Graph

import pandas as pd

# never used subgrounds so getting used to the syntax here
```

```python
# initialize Subgrounds object
sg = Subgrounds()

# load livepeer subgraph
lpt = sg.load_subgraph('https://api.thegraph.com/subgraphs/name/livepeer/livepeer')
```

```python
# define FieldPaths
transcoders = lpt.Query.transcoders
delegators = lpt.Query.delegators
```

```python
top_10_transcoders = transcoders(
    orderBy=transcoders.totalStake,
    orderDirection='desc',
    first=10,
    where=[
        transcoders.active == True
    ]
)

sg.query_df([
    top_10_transcoders.id,
    top_10_transcoders.totalStake,
    top_10_transcoders.totalVolumeETH,
    top_10_transcoders.pricePerSegment
])
```

```python
# call ratio
active_transcoders = transcoders(
    orderBy=transcoders.totalStake,
    orderDirection='desc',
    first=10,
    where=[
        transcoders.active == True
    ]
)

# subgrounds does not really account for null as an option
# solution is to grab the last 30 non-null and max-min will tell us how many were null
active_pools = active_transcoders.pools(
    first=30,
    where=[
        active_transcoders.pools.rewardTokens != None
    ]
)

pool_rewards = sg.query_df([
    active_transcoders.id,
    active_pools.round.id,
    active_pools.rewardTokens
])

pool_rewards

# req = sg.mk_request([
#     active_transcoders.id,
#     active_pools.round.id,
#     active_pools.rewardTokens
# ])

# print(req.graphql)
```

```python
grouped = pool_rewards.groupby('transcoders_id')
rmin = grouped.min()['transcoders_pools_round_id'].rename('min').astype('int64')
rmax = grouped.max()['transcoders_pools_round_id'].rename('max').astype('int64')
```

```python
top_transcoders = pd.concat([rmin,rmax],axis=1).reset_index()
```

```python
# i'd much rather count null values than go thru this, but they're working on that bug
top_transcoders['diff'] = top_transcoders['max'] - top_transcoders['min']
# inclusive so 29 = 30 rounds
top_transcoders['missing'] = top_transcoders['diff'] - 29
top_transcoders['implied_rate'] = 1 - (top_transcoders['missing'] / 30)
top_transcoders
```
